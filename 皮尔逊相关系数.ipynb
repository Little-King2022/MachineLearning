```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_simple = np.array([-2, -1, 0, 1, 2])
y_simple = np.array([4, 1, 3, 2, 0])
my_rho = np.corrcoef(x_simple, y_simple)  # corrcoef 相关系数

print(my_rho)
```

```python
# 从此开始

seed = 13
rand = np.random.RandomState(seed)
```

```python
# random.uniform(x, y)方法将随机生成一个实数，它在 [x,y] 范围内。

x = rand.uniform(0,1,100)  # 第一次rand.uniform()调用会生成随机均匀分布
```

```python
print(x)
```

```python
x = np.vstack((x, x*2+1))  # 竖直堆叠序列中的数组（行方向） 调用vstack()将其他数组垂直堆叠到该数组
```

```python
print(x) # 第二个变量与第一个变量具有完全正相关. 因为第二个变量是在第一个变量的基础上堆叠的一些变量集
```

```python
x.shape
```

```python
x = np.vstack((x, -x[0,]*2+1))  # 第三个变量与第一个变量具有完全负相关
```

```python
#  np.random.normal()正态分布
#  函数的原型np.random.normal(loc,scale,size)，其中loc表示均值，scale表示方差，size表示输出的size
#  功能：该函数用于生成高斯随机分布是随机数

x = np.vstack((x, rand.normal(1,3,100)))  # 第四个变量是完全随机的
```

```python
print(x)
```

```python
x.shape
```

```python
rho = np.corrcoef(x)
print(rho)
```

```python
fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(12, 3))  # （fig：是figure的缩写）
```

```python
fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(12, 3))  # （fig：是figure的缩写）
for i in [0,1,2]:
    
    # 第一个绘图区域的横坐标描绘x里的第一个元素, 纵坐标描绘x里的第二个元素
    ax[i].scatter(x[0,],x[1+i,])  
    
    # rho表示x的皮尔逊相关系数, 第一个图是第0行第一列 第二个图是第0行第二列
    ax[i].title.set_text('Correlation = ' + "{:.2f}".format(rho[0,i+1]))  
    
    ax[i].set(xlabel='x',ylabel='y')
    
fig.subplots_adjust(wspace=.4)    
plt.show()
```

```python
fig, ax = plt.subplots(nrows=2, ncols=4, figsize=(15, 8))

for noise, i in zip([0.05,0.2,0.8,2],[0,1,2,3]):
    # Add noise
    x_with_noise = x + rand.normal(0,noise,x.shape)
    
    # Compute correlation
    rho_noise = np.corrcoef(x_with_noise)
    
    # Plot column wise. Positive correlation in row 0 and negative in row 1
    ax[0,i].scatter(x_with_noise[0,],x_with_noise[1,],color='magenta')
    ax[1,i].scatter(x_with_noise[0,],x_with_noise[2,],color='green')
    ax[0,i].title.set_text('Correlation = ' + "{:.2f}".format(rho_noise[0,1])
                        + '\n Noise = ' + "{:.2f}".format(noise) )
    ax[1,i].title.set_text('Correlation = ' + "{:.2f}".format(rho_noise[0,2])
                        + '\n Noise = ' + "{:.2f}".format(noise))
    ax[0,i].set(xlabel='x',ylabel='y')    
    ax[1,i].set(xlabel='x',ylabel='y')
    
fig.subplots_adjust(wspace=0.3,hspace=0.4)    
plt.show()
```
